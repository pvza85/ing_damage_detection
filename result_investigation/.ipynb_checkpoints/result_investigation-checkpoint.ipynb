```python
import pickle
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import fbeta_score
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')

train_pred = pickle.load(open('../results/train_pred.pik', 'r'))
test_pred = pickle.load(open('../results/test_pred.pik', 'r'))

train_pred = np.argmax(train_pred, axis=1)
test_pred = np.argmax(test_pred, axis=1)

df = pd.read_csv('../data/current/labeling.csv')

train = df.head(len(train_pred))
test = df.tail(len(test_pred))
```

```python
import datetime
def gettime():
    now = datetime.datetime.now().time()
    return str(now.hour) + '_' + str(now.minute)
print gettime()
```

```python
target = train['manual_label']
pred = train_pred

print 'accuracy_score: ', accuracy_score(target, pred)
print 'recall_score: ', recall_score(target, pred)
print 'precision_score: ', precision_score(target, pred)
print 'f2_score: ', fbeta_score(target, pred, 2)
print 'confusion_matrix: '
print confusion_matrix(target, pred)
```

```python
target = test['manual_label']
pred = test_pred

print 'accuracy_score: ', accuracy_score(target, pred)
print 'recall_score: ', recall_score(target, pred)
print 'precision_score: ', precision_score(target, pred)
print 'f2_score: ', fbeta_score(target, pred, 2)
print 'confusion_matrix: '
print confusion_matrix(target, pred)
```
